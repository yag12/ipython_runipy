```python
p = apachelog.parser(fformat)
```

```python
log = open('/var/log/apache2/yts.lo-access.log').readlines()
```

```python
log_list = []
for line in log:
       try:
          data = p.parse(line)
       except:
          sys.stderr.write("Unable to parse %s" % line)
                
       data['%t'] = data['%t'][1:12]+' '+data['%t'][13:21]+' '+data['%t'][22:27]
      
       log_list.append(data)
```

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame, Panel
```

```python
df = DataFrame(log_list)
```

```python
df[0:2]
```

```python
del df['%l']; del df['%u']; del df['%i'];
```

```python
df = df.rename(columns={'%>s': 'Status', '%b':'KB', '%h':'IP', '%r':'Request', '%t': 'Time', '%d':'Date', '%{User-Agent}i':'Device'})
```

```python
df.head()
```

```python
df.index = pd.to_datetime(df.pop('Time'))
```

```python
df['Status'] = df['Status'].astype('int')
```

```python
def dash2nan(x):
    if x == '-':
        x = np.nan
    else:
        x = float(x)/1024.
    
    return x
```

```python
df['KB'] = df['KB'].apply(dash2nan)
```

```python
df['KB'].plot()
```

```python
df.head()
```

```python
t_span = '1D'
df_date = DataFrame({'Access':df['Status'][df['Status'] != 404].resample(t_span, how='count'),'KB':df['KB'].resample(t_span, how='sum'),})
```

```python
df_date
```

```python
df_date.plot()
```
